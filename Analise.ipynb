```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

%matplotlib inline
```

```python
df = pd.read_csv('cadeiras.csv')
```

# Distribuição da dificuldade das cadeíras por período

```python
fig, ax = plt.subplots()

fig.set_size_inches(10,5)
ax.set_title('Curva de dificuldade do curso por períodos')
ax.set(xlabel='Período', ylabel='Média de 1-5',
       title='Dificuldade das cadeiras por período')
ax.grid(alpha=0.25)
ax.set_ylim(top=5,bottom=1)

ax.scatter(df['Período'], df.mean(axis=1),color='crimson')
```

```python
dificuldades = []
periodos = ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8', 'P9']

for periodo in periodos:
    dif = df[df['Período'] == periodo].mean(axis=1).sum(axis=0)
    d = round(dif, 2)
    dificuldades.append(d)
```

# Dificuldade Geral do Período

```python
fig, ax = plt.subplots()

fig.set_size_inches(10,5)
ax.set_title('Curva de dificuldade do curso por períodos')
ax.set(ylabel='Soma das médias das dificuldades',
       title='Dificuldade por período')
ax.grid(alpha=0.35)
ax.set_ylim(top=25,bottom=3)

ax.plot(periodos, dificuldades)
```

## Média de dificuldade por cadeira

```python
df['Média'] = df.mean(axis=1)
df = df[['Período', 'Cadeira', 'Média']]
df['Média'] = round(df['Média'], 2)

df.sort_values(by='Média', ascending=False)
```
