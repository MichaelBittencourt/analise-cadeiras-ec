```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

%matplotlib inline
```

```python
df = pd.read_csv('cadeiras.csv')
```

# Distribuição da dificuldade das cadeíras por período

```python
fig, ax = plt.subplots()

fig.set_size_inches(10,5)
ax.set_title('Curva de dificuldade do curso por períodos')
ax.set(xlabel='Período', ylabel='Média de 1-5',
       title='Dificuldade das cadeiras por período')
ax.grid(alpha=0.25)
ax.set_ylim(top=5,bottom=1)

ax.scatter(df['Período'], df.mean(axis=1),color='crimson')
```

```python
dificuldades = []
periodos = ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8', 'P9']

for periodo in periodos:
    dif = df[df['Período'] == periodo].mean(axis=1).sum(axis=0)
    d = round(dif, 2)
    dificuldades.append(d)
```

# Dificuldade Geral do Período

```python
fig, ax = plt.subplots()

fig.set_size_inches(10,5)
ax.set_title('Curva de dificuldade do curso por períodos')
ax.set(ylabel='Soma das médias das dificuldades',
       title='Dificuldade por período')
ax.grid(alpha=0.35)
ax.set_ylim(top=25,bottom=3)

ax.plot(periodos, dificuldades)
```

## Média de dificuldade por cadeira

```python
df['Média'] = df.mean(axis=1)
df = df[['Período', 'Cadeira', 'Média']]
df['Média'] = round(df['Média'], 2)

df.sort_values(by='Média', ascending=False)
```

## Média de dificuldade por período

```python
df1 = df[df['Período'] == 'P1']
df2 = df[df['Período'] == 'P2']
df3 = df[df['Período'] == 'P3']
df4 = df[df['Período'] == 'P4']
df5 = df[df['Período'] == 'P5']
df6 = df[df['Período'] == 'P6']
df7 = df[df['Período'] == 'P7']
df8 = df[df['Período'] == 'P8']
df9 = df[df['Período'] == 'P9']

data = {}
data ['P1'] = [df1['Média'].sum()/df1.shape[0]]
data ['P2'] = [df2['Média'].sum()/df2.shape[0]] 
data ['P3'] = [df3['Média'].sum()/df3.shape[0]]
data ['P4'] = [df4['Média'].sum()/df4.shape[0]]
data ['P5'] = [df5['Média'].sum()/df5.shape[0]] 
data ['P6'] = [df6['Média'].sum()/df6.shape[0]] 
data ['P7'] = [df7['Média'].sum()/df7.shape[0]] 
data ['P8'] = [df8['Média'].sum()/df8.shape[0]] 
data ['P9'] = [df9['Média'].sum()/df9.shape[0]] 

df_periodo = pd.DataFrame(data=data)

df_periodo = df_periodo.T
df_periodo.index.name = 'Período'
df_periodo.columns = ['Média']
df_periodo['Média'] = round(df_periodo['Média'], 2)

df_periodo
```
